```python
from scipy.io import loadmat
import pandas as pd
import altair as alt
import numpy as np
alt.data_transformers.enable('data_server')
```

```python
gl_10 = loadmat('../data/raw/gl_10.mat')
```

```python
signal = pd.DataFrame(gl_10["SIG"][0][1]).T
```

```python
signal = signal.reset_index().rename(columns={"index": "time", 0:"emg"})
```

```python
signal["time"] = signal["time"] / 2048
```

```python
signal
```

```python
alt.Chart(signal, title="Raw EMG data from single channel").encode(
    x=alt.X("time", title="Time (s)", scale=alt.Scale(nice=False)),
    y=alt.Y("emg", title="Voltage", axis=alt.Axis(labels=False, ticks=False)),
).mark_line(strokeWidth=0.2).configure_axis(
    grid=False, labelFontSize=14, titleFontSize=20
).configure_view(strokeOpacity=0).configure_title(fontSize=20)
```
